GET- It is used when request for data from server.
characterstics of GET method:
1. Request parameter- Data which is requested is send as parameter into url as query string.
2. Vsisbility- Since data is seent as parameter it is clearly visible on address bar
3. Data limit- Get request has limit of data which user can request which is imposed by browser or server


POST- It is a method of communication which is used when the user want to send data to server

characterstics of POST:
    
1.Request parameter- Data is sent from request body not visible in url
2. Data Volume- Post request can handle larger amounts of data compared to get requsts. Suitable for sending and uploading files
3. Safe- It is considered more secure since data is not exposed to url
